```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import math
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split


data = pd.read_csv('train.csv', usecols = [1,2,4,5,6,7])

data.head()

#replace NA values
data['Age'].fillna(0, inplace=True)
```

```python
data = pd.get_dummies(data, drop_first = True)
data.sample(50)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(data.drop('Survived', axis=1), data['Survived'])
```

```python
print(xtrain)
reg = LogisticRegression(solver = 'saga')
reg.fit(xtrain, ytrain)
```

```python
#read test csv file

test = pd.read_csv('test.csv', usecols = [1,3,4,5,6])

test.head()

test['Age'].fillna(0, inplace=True)

test = pd.get_dummies(test, drop_first = True)
test.sample(50)

print(test.loc[0])
```

```python
#create csv file
import csv

with open("titanic_submission.csv", mode = "w") as csvfile:
  fieldnames = ["PassengerId", "Survived"]
  writer = csv.DictWriter(csvfile, fieldnames = fieldnames)
  writer.writeheader()
  guess = 0
  for i in range(test['Age'].size):
    guess = reg.predict([test.loc[i]])[0]
    writer.writerow({"PassengerId": 892 + i, "Survived": guess})

```
